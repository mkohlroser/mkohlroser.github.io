First Lets import our data and inspect. We are workging with two quantitative, continuous variables GDP (USD) and Internet Penetration (%). 

```python
import pandas as pd
```

```python
df_GDP = pd.read_csv("GDP.csv", encoding="utf-8")

df_GDP.head()
```

```python
df_INT = pd.read_csv("Internet.csv", encoding="utf-8")

df_INT.columns

df_INT.head()
```

Yuck wide form data, Lets convert this to long form. This will be more useful for visualizing the relationship betweem the two variables.

```python
# Convert GDP Data to Long Form 

long_GDPdf = df_GDP.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="GDP"
)

long_GDPdf["Year"] = pd.to_numeric(long_GDPdf["Year"], errors="coerce")

long_GDPdf.head()

long_GDPdf["GDP"].describe()

long_GDPdf["Country Name"].unique

```

```python
# Convert INT Data to Long Form 

long_INTdf = df_INT.melt(
    id_vars = ["Country Name", "Country Code", "Indicator Name"],
    var_name = "Year", 
    value_name = "Internet Penetration"
)

long_INTdf["Year"] = pd.to_numeric(long_INTdf["Year"], errors="coerce")

long_INTdf.head()
```

By converting to long form we have created a data set that shows the change in GDP/INT penetration overtime by country. We now have two new columns "GDP" and "Internet Penetration" that store the values of the observation - that is whatever the GDP/ Internet Penetration is for a specific country at a given year. 

```python
merged_df = pd.merge(long_GDPdf, long_INTdf, on=["Country Name", "Country Code", "Year"], how="left")

```

Now that we have ensured they are both in long from, we can easily merge them by Country Name, Country Code, and Year which should be constant across both datasets. 

```python
merged_df.head()
```

This looks like it worked, but it seems silly to include years prior to 1990. We can clean this further

```python
# Keep Years greater than 1990 (this is why we converted Year to a numeric)

merged_df_clean = merged_df[merged_df["Year"] > 1990]

merged_df_clean.head()
```

With that done I think we are ready for our first plot! 

```python
import plotly.express as px

fig = px.scatter(merged_df_clean, x="Internet Penetration", y="GDP")
fig.show()
```

Hmmm if figured this would happen. GDP numbers are so large it messes with how it is visualized. Also, what's going on with the x axis? Let's try taking the log of GDP and see how that looks. 

```python
import numpy as np

merged_df_clean["ln_gdp"] = np.log(merged_df_clean["GDP"])

merged_df_clean["ln_gdp"].head(20)
```

```python
fig2 = px.scatter(merged_df_clean, y="ln_gdp", x="Internet Penetration",  trendline="ols")
fig2.update_layout(title = "Internet Penetration Vs GDP")
fig2.data[1].line.color = "red"
fig2.show()
```

That's much better! Although I still have some concerns about the GDP values.

It looks like internet penetration and ln_GDP have a weak positive correlation. Lets double check that. 

```python
round(merged_df_clean["GDP"].corr(merged_df_clean["Internet Penetration"]),3)
```

As Internet Penetration Increases so does GDP. These two variables display a weak, positive, and linear relationship. Evidenced by our scatterplot and our correlation statistic of .151.  But what's with these massive GDP values? 

```python
merged_df_clean["Country Name"].unique
```

Ahh I see its including country groupings also. Thats why this is so messed up. 

```python
merged_df_clean = merged_df_clean[~merged_df_clean["Country Name"].str.contains(" ")]

to_remove = ["World", 
             "Asia", "Africa", "Europe", "North America", 
             "Aouth America", "Antarctica", "Australia"]

merged_df_clean = merged_df_clean[~merged_df_clean["Country Name"].isin(to_remove)]


merged_df_clean["Country Name"].unique

```

```python
fig2 = px.scatter(merged_df_clean, y="ln_gdp", x="Internet Penetration",  trendline="ols")
fig2.update_layout(title = "Internet Penetration Vs GDP")
fig2.data[1].line.color = "red"
fig2.show()

round(merged_df_clean["GDP"].corr(merged_df_clean["Internet Penetration"]),3)
```

After cleaning the data further our r statistic increased to .194. If we continue to clean the data that number will change again!
